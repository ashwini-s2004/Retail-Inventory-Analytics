## Objective

```python
category_count = df.groupby('Category')['Product ID'].count()

category_count.sort_values(ascending=False)
```

This analysis shows the number of products available in each category and helps identify categories with larger inventory presence.

This analysis identifies the categories contributing the highest sales volume and highlights fast-moving inventory segments.

### Category-wise Revenue Analysis

Revenue contribution by category helps management identify the most profitable product segments.

```python
top_products = df.sort_values(
    by='Units Sold',
    ascending=False
).head(10)

top_products[['Product Name',
              'Category',
              'Units Sold']]
```

### Top 10 Best-Selling Products

These products recorded the highest sales volumes and represent the most demanded inventory items.

```python
top_revenue_products = df.sort_values(
    by='Revenue',
    ascending=False
).head(10)

top_revenue_products[['Product Name',
                      'Revenue']]
```

### Top Revenue Products

These products generate the highest revenue and contribute significantly to business profitability.

```python
df['Total Profit'] = df['Profit Per Unit'] * df['Units Sold']

top_profit_products = df.sort_values(
    by='Total Profit',
    ascending=False
).head(10)

top_profit_products[['Product Name',
                     'Total Profit']]
```

# 📈 Task 4: Data Visualization

## Objective

## 📊 Visualization 1: Category-wise Units Sold

```python
category_sales = df.groupby('Category')['Units Sold'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))

category_sales.plot(kind='bar')

plt.title("Category-wise Units Sold")
plt.xlabel("Category")
plt.ylabel("Units Sold")

plt.xticks(rotation=45)

plt.show()
```

### Observation

The chart shows the total units sold by each category.

Categories with higher sales indicate strong customer demand and require better inventory planning.

## 🥧 Visualization 2: Product Category Distribution

```python
category_count = df['Category'].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    category_count,
    labels=category_count.index,
    autopct='%1.1f%%'
)

plt.title("Product Category Distribution")

plt.show()
```

### Observation

The pie chart displays the percentage contribution of each product category.

This helps understand inventory allocation across categories.

## 📊 Visualization 3: Units Sold Distribution

```python
plt.figure(figsize=(10,6))

plt.hist(df['Units Sold'], bins=20)

plt.title("Distribution of Units Sold")

plt.xlabel("Units Sold")

plt.ylabel("Frequency")

plt.show()
```

### Observation

The histogram shows how frequently different sales quantities occur.

It helps identify whether most products have low, medium, or high sales.

## 📈 Visualization 4: Opening Stock vs Closing Stock

```python
plt.figure(figsize=(12,6))

plt.plot(df['Opening Stock'])

plt.plot(df['Closing Stock'])

plt.title("Opening Stock vs Closing Stock")

plt.xlabel("Product Index")

plt.ylabel("Stock Quantity")

plt.legend([
    'Opening Stock',
    'Closing Stock'
])

plt.show()
```

### Observation

The line graph compares opening stock and closing stock levels.

Large gaps indicate significant product movement and inventory consumption.

## 🔥 Visualization 5: Correlation Heatmap

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

The heatmap shows relationships among numerical variables.

Strong positive or negative correlations help identify key inventory and sales drivers.

## 🏆 Top 10 Best Selling Products

```python
top_products = df.sort_values(
    by='Units Sold',
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x='Units Sold',
    y='Product Name',
    data=top_products
)

plt.title("Top 10 Best Selling Products")

plt.show()
```

# 💡 Task 5: Business Insights & Recommendations

## Objective

The purpose of this analysis is to derive actionable business insights from inventory data and provide recommendations that can improve inventory management, reduce stock-out risks, increase sales efficiency, and optimize supplier relationships.

## 🔴 Insight 1: Products Requiring Immediate Replenishment

```python
reorder_products = df[df['Closing Stock'] <= df['Reorder Point']]

reorder_products[['Product Name',
                  'Closing Stock',
                  'Reorder Point']].head(10)
```

### Business Insight

Several products have reached or fallen below their reorder points.

These products are at high risk of stock shortages and may lead to lost sales opportunities if replenishment is delayed.

### Recommendation

Implement automatic reorder alerts and maintain safety stock levels for critical products.

## 🟢 Insight 2: Fast-Moving Product Categories

```python
category_sales = df.groupby('Category')['Units Sold'].sum()

category_sales.sort_values(ascending=False)
```

### Business Insight

The category with the highest units sold represents the fastest-moving inventory segment.

These categories contribute significantly to sales performance and customer demand.

### Recommendation

Maintain higher inventory levels and prioritize replenishment for fast-moving categories.

## 🟡 Insight 3: Slow-Moving Products

```python
slow_products = df.sort_values(
    by='Units Sold'
).head(10)

slow_products[['Product Name',
               'Units Sold']]
```

### Business Insight

Certain products exhibit very low sales volume and occupy storage space for extended periods.

These products increase inventory carrying costs and reduce warehouse efficiency.

### Recommendation

Consider discount strategies, promotional campaigns, or inventory reduction plans for slow-moving items.

## 💰 Insight 4: High-Profit Products

```python
df['Total Profit'] = (
    df['Retail Price']
    -
    df['Unit Cost']
) * df['Units Sold']

high_profit = df.sort_values(
    by='Total Profit',
    ascending=False
).head(10)

high_profit[['Product Name',
             'Total Profit']]
```

### Business Insight

Several products generate significantly higher profits compared to others.

These products contribute strongly to overall business profitability.

### Recommendation

Increase marketing focus, stock availability, and supplier support for high-profit products.

## 🚚 Insight 5: Supplier Dependency Risk

```python
supplier_products = df.groupby(
    'Supplier Name'
)['Product ID'].count()

supplier_products.sort_values(
    ascending=False
)
```

### Business Insight

A large percentage of products are supplied by a limited number of suppliers.

Heavy dependence on a few suppliers may create supply chain risks during disruptions.

### Recommendation

Diversify supplier partnerships and establish backup suppliers for critical inventory categories.

## 📈 Bonus Insight: Revenue Contribution Analysis

```python
category_revenue = df.groupby(
    'Category'
)['Revenue'].sum()

category_revenue.sort_values(
    ascending=False
)
```

### Business Insight

Certain categories contribute disproportionately to total revenue.

These categories represent key growth opportunities for the business.

### Recommendation

Allocate marketing and inventory resources toward high-revenue categories.

# 📋 Final Inventory Management Recommendations

Based on the analysis, the following recommendations are proposed:

### 1. Implement Automated Reorder Alerts

Products approaching reorder points should trigger automatic notifications.

### 2. Maintain Safety Stock

High-demand products should have additional buffer inventory to avoid stock-outs.

### 3. Reduce Slow-Moving Inventory

Low-demand products should be reviewed regularly and optimized.

### 4. Focus on High-Profit Products

Increase inventory and marketing support for high-margin items.

### 5. Diversify Suppliers

Reduce dependency on individual suppliers by establishing alternative sourcing options.

### 6. Monitor Category Performance Monthly

Regular sales and inventory reviews can improve forecasting accuracy.
